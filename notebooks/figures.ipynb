## Overlap fraction and 3D distance comparison:
### Example of Fig.1C, i.e. 51a-51b (within TAD) versus 51b-52 (between TADs) in ESC; see graph 1c.

```python
# Loading data
esc_df = pd.read_csv("../assays/ESC/Experiment_A_merged_df.csv")
npc_df = pd.read_csv("../assays/NPC/Experiment_B_merged_df.csv")
ncxnpc_df = pd.read_csv("../assays/ncxNPC/Experiment_C_merged_df.csv")
ctcf_aid_ctl_df = pd.read_csv("../assays/CTCF-AID_AUX-CTL/Experiment_D_merged_df.csv")
ctcf_aid_aux_df = pd.read_csv("../assays/CTCF-AID_AUX/Experiment_E_merged_df.csv")
rad21_aid_ctl_df = pd.read_csv("../assays/RAD21-AID_AUX-CTL/Experiment_F_merged_df.csv")
rad21_aid_aux_df = pd.read_csv("../assays/RAD21-AID_AUX/Experiment_G_merged_df.csv")
esc_ctl_df = pd.read_csv("../assays/ESC_TSA-CTL/Experiment_H_merged_df.csv")
esc_tsa_df = pd.read_csv("../assays/ESC_TSA/Experiment_I_merged_df.csv")
```

```python
# Selecting some data
probes_sel = ["51a", "51b", "52"]
esc_subset = esc_df[(esc_df['Probe Ch0'].isin(probes_sel)) & (esc_df['Probe Ch1'].isin(probes_sel))]
esc_subset.head()
```

```python
# Plotting overlap fraction
esc_subset.boxplot(column="overlap_fraction", by=["Probe Ch0", "Probe Ch1"])
```

```python
# Plotting 3D-distance
esc_subset.boxplot(column="distance3d", by=["Probe Ch0", "Probe Ch1"])
```

### Example of Fig.2, i.e. 21-22 in ESC (split TAD) versus 21-22 in NPC (merge TAD); see graph 2g.

```python
probes_sel = ["21", "22", 21, 22]
esc_subset = esc_df[(esc_df['Probe Ch0'].isin(probes_sel)) & (esc_df['Probe Ch1'].isin(probes_sel))]
esc_subset.head()
```

```python
npc_subset = npc_df[(npc_df['Probe Ch0'].isin(probes_sel)) & (npc_df['Probe Ch1'].isin(probes_sel))]
```

```python
ncxnpc_subset = ncxnpc_df[(ncxnpc_df['Probe Ch0'].isin(probes_sel)) & (ncxnpc_df['Probe Ch1'].isin(probes_sel))]
ncxnpc_subset.head()
```

```python
# Plotting 3D-distance
pd.concat([esc_subset, npc_subset, ncxnpc_subset]).boxplot(column="distance3d",
                                                           by=["Cell Line Name", "Probe Ch0", "Probe Ch1"])
```

### Examples of Fig.3: 101-102a (between TADs) for CTCF-AID-control versus CTCF-AID-AUX; and 102a-102b (within TAD) for RAD21-AID-control versus RAD21-AID-AUX; see graph 3c.

```python
probes_sel = ["101", "102a", "102b"]

ctcf_aid_ctl_subset = ctcf_aid_ctl_df[(ctcf_aid_ctl_df['Probe Ch0'].isin(probes_sel)) & (ctcf_aid_ctl_df['Probe Ch1'].isin(probes_sel))]
ctcf_aid_ctl_subset.head()
ctcf_aid_aux_subset = ctcf_aid_aux_df[
    (ctcf_aid_aux_df['Probe Ch0'].isin(probes_sel)) & (ctcf_aid_aux_df['Probe Ch1'].isin(probes_sel))]
ctcf_aid_aux_subset.head()

rad21_aid_ctl_subset = rad21_aid_ctl_df[
    (rad21_aid_ctl_df['Probe Ch0'].isin(probes_sel)) & (rad21_aid_ctl_df['Probe Ch1'].isin(probes_sel))]
rad21_aid_ctl_subset.head()
rad21_aid_aux_subset = rad21_aid_aux_df[
    (rad21_aid_aux_df['Probe Ch0'].isin(probes_sel)) & (rad21_aid_aux_df['Probe Ch1'].isin(probes_sel))]
rad21_aid_aux_subset.head()
```

```python
pd.concat([ctcf_aid_ctl_subset, ctcf_aid_aux_subset]).boxplot(column="overlap_fraction", by=["Protocol", "Probe Ch0", "Probe Ch1"], rot=45)

pd.concat([rad21_aid_ctl_subset, rad21_aid_aux_subset]).boxplot(column="overlap_fraction", by=["Protocol", "Probe Ch0", "Probe Ch1"], rot=45)

pd.concat([ctcf_aid_ctl_subset, ctcf_aid_aux_subset]).boxplot(column="distance3d", by=["Protocol", "Probe Ch0", "Probe Ch1"], rot=45)

pd.concat([rad21_aid_ctl_subset, rad21_aid_aux_subset]).boxplot(column="distance3d", by=["Protocol", "Probe Ch0", "Probe Ch1"], rot=45)
```

## CND/subdomain number comparisons:
